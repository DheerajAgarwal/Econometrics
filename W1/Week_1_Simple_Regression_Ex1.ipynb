```python
file_location=".\\data\\TrainExer11.txt"
ex1_data=pd.read_csv(file_location, sep= "\t")
ex1_data.head()
```

```python
import bkcharts as bc
from bokeh.io import output_notebook
from bokeh.plotting import show
from bokeh.layouts import row

output_notebook()
```

```python
hist_age = bc.Histogram(ex1_data, 
                        values="Age", 
                        title = "Freq Dist of Age", 
                        plot_width=450, 
                        plot_height = 400)
hist_expense = bc.Histogram(ex1_data, 
                            values="Expenditures", 
                            title = "Freq Dist of Expense", 
                            plot_width=450, 
                            plot_height = 400)
```

```python
show(row(hist_age, hist_expense))
```

```python
age_expense_scatter = bc.Scatter(ex1_data, 
                                 x="Age", 
                                 y="Expenditures", 
                                 title = "Expense vs. Age", 
                                 plot_width = 900, 
                                 plot_height = 300)
```

```python
show(age_expense_scatter)
```

We can observe that on the whole the scatter plot of age vs expenditure does not demonstrate the clear linear relationship which the sales and the price data in the lecture demonstrated.

However, if we look at the data as two separate segments, segment 1 under the age of 40 and segment 2, age 40 or above, we can start seeing that there are relationships that exists within each segments. We also note that the expense increases slightly with increase in age within segment 1, howver its hard to see any trend within segment 2.

```python
sample_mean_expense = ex1_data.Expenditures.mean()
round(sample_mean_expense, 1)
```

Compute two sample means of expenditures, one for clients of age 40 or more and the other for clients of age below 40.

What daily expenditures would you predict for a new client of 50 years old? And for someone who is 25 years old?

```python
msk = ex1_data.Age < 40
print("Ave Expense for people under 40 is: ", round(ex1_data.Expenditures[msk].mean(), 1))

msk = ex1_data.Age >= 40
print("Ave Expense for people older or equal to 40 is: ", round(ex1_data.Expenditures[msk].mean(), 1))
```

Based on the a simplistic approach, the sample mean is the predicted value.
